```python
import networkx as nx
from networkx import *
import matplotlib.pyplot as plt
```

# Cyclic

```python
G = nx.Graph()
elist = [('A', 'B'), ('B', 'C'), ('C', 'A')]
labels = dict()
labels[('A','B')]=1
labels[('B','C')]=2
labels[('C','A')]=3
G.add_edges_from(elist)
nx.draw_networkx_labels(G, pos=nx.circular_layout(G))
nx.draw_networkx_edge_labels(G, pos=nx.circular_layout(G), edge_labels=labels)
nx.draw_networkx(G, pos = circular_layout(G), node_color='violet')
plt.savefig('Cyclic.pdf',bbox_inches='tight')
```

# L(Cyclic)

```python
G = nx.Graph()
elist = [('1', '2'), ('2', '3'), ('3', '1')]
labels = dict()
labels[('1','2')]='A'
labels[('2','3')]='B'
labels[('3','1')]='C'
G.add_edges_from(elist)
nx.draw_networkx_labels(G, pos=nx.circular_layout(G))
nx.draw_networkx_edge_labels(G, pos=nx.circular_layout(G), edge_labels=labels)
nx.draw_networkx(G, pos = circular_layout(G), node_color='springgreen')
plt.savefig('LCyclic.pdf',bbox_inches='tight')
```
